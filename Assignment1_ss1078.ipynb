(a) Vectorize this code and demonstrate (as in the last exercise) the speed increase through vectorization and (b) plot the resulting data - both the function $f(x,y)$ and the thresholded output - using [`imshow`](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.imshow.html?highlight=matplotlib%20pyplot%20imshow#matplotlib.pyplot.imshow) from `matplotlib`.

*(a)* Vectorize this code and demonstrate (as in the last exercise) the speed increase through vectorization

```python
# import numpy as np
import time
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

nvalues = 2000
xvalues = np.linspace(-4,4,nvalues)
yvalues = np.linspace(-4,4,nvalues)
thresh  = 0

# Nonvectorized implementation
t0 = time.time()
f = np.zeros((nvalues,nvalues))
f_thresholded = np.zeros((nvalues,nvalues))
for ix, x in enumerate(xvalues):
    for iy, y in enumerate(yvalues):
        f[ix,iy]             = x**2 - 2 * y**2
        f_thresholded[ix,iy] = f[ix,iy] > thresh
t1 = time.time()
time_nonvectorized = t1 - t0
print("Time [sec] (non-vectorized):" ,round(time_nonvectorized,4))

# Vectorized implementation
t2 = time.time()
## combining x and y into a matrix with meshgrid
xv, yv = np.meshgrid(xvalues, yvalues)
## applying required function: (ð‘¥,ð‘¦)=ð‘¥^2âˆ’2ð‘¦*2
z = xv**2 - (2 * yv**2)
## transpose to get the result in the desired format
z=z.transpose()
## applying check condition to check if result is above a threshold = 0
z_thresholded= z>thresh
## converting resultant thresholded matrix into integers
#z_thresholded=z_thresholded.astype(int) 
t3 = time.time()

# Vectorization speed performance results
time_vectorized = t3 - t2
print("Time [sec] (vectorized):" ,round(time_vectorized,4))
print("The vectorized code is", round(time_nonvectorized/time_vectorized,2) ,"times faster than the vectorized code, or the speed is increased by ",round(((time_nonvectorized-time_vectorized)/time_nonvectorized)*100,2),"%.")
```

### Investigation of Target Variable (distribution)

#### What are we doing?
We are investigating the nature of the target or response variable here; i.e. 'SalePrice'. On the basis of our findings here, we can transform the variable.

#### Why are we doing this?
We are investigating and transforming variables in this step to ensure there is no evident heteroscedasticity i.e. non-uniform spread of residuals over the range of the different target values. This is to ensure the model predictions behave better. What this means is, in regression it is key that the residuals follow a normal distribution and are not skewed towards the smaller or the larger predictions. 

```python
# investigating the basic stats of 'SalePrice'
train_df['SalePrice'].describe()
# here we see min= 34900 and max= 625000
# to investigate further and derive more insights, we will plot a distribution curve
#distribution plot- histogram
sns.distplot(train_df['SalePrice']).set_title("Distribution of SalePrice")

# probability plot
fig = plt.figure()
res = stats.probplot(train_df['SalePrice'], plot=plt)
```

We see a positive skewness. This is not exactly a normal distribution. Also, the balues do not follow the linear trend here.  
Now, if we observe a positive or a right skewness, log transformation is a good option

#### Why log?
The non-linear trend shows 'SalePrice' has some sort of exponential relationship with the independent variables. Applying a log function on these values should give a linear trend and convert the set of values into 'normally distributed' values.

```python
#Use the numpy fuction log1p which  applies log(1+x) to all elements of the column
train_df["SalePrice"] = np.log1p(train_df["SalePrice"])

#new distribution 
sns.distplot(train_df['SalePrice'] , fit=norm);

# Get the mean and std deviations
(mu, sigma) = norm.fit(train_df['SalePrice'])
print( '\n mean = {:.3f} and std dev = {:.3f}\n'.format(mu, sigma))

#distribution
plt.legend(['Normal dist. ($\mu=$ {:.3f} and $\sigma=$ {:.3f} )'.format(mu, sigma)],
            loc='best')
plt.ylabel('Frequency')
plt.title('Distribution of Log SalePrices')

#QQ-plot
fig = plt.figure()
res = stats.probplot(train_df['SalePrice'], plot=plt)
plt.show()
```

Since the distribution of the target variable has been made normal, we should also look into some of the highly correlated features. This will ensure the assumption of normality is followed whicle creating most basic regression models.

We cannot simply impute this column with the mode or average as the column Lot Frontage area indicates an area figure. This is a continuous variable that might be dependent on other variables defining the attributes of the house. For instance, the street it is on, the area of the house etc.

We will use a slightly more intuitive model to accommodate for these factors. We will use a simple regression model to impute LotFrontage

```python
# function to scale a column
def norm_minmax(col):
    return (col-col.min())/(col.max()-col.min())
```

```python
# By business definition, LotFrontage is the area of each street connected to the house property
#creating a model by removing SalePrice and with other features

# convert categoricals to dummy varables
#also dropping the target 'SalePrice' for now, as the target in this model is 'LotFrontage'
total_df_dummy = pd.get_dummies(total_df.drop('SalePrice',axis=1))
# scaling all numerical columns
for col in total_df_dummy.drop('LotFrontage',axis=1).columns:
    total_df_dummy[col] = norm_minmax(total_df_dummy[col])

frontage_train = total_df_dummy.dropna()

#define y and x variables
frontage_train_y = frontage_train.LotFrontage
frontage_train_X = frontage_train.drop('LotFrontage',axis=1)  

# define and fit model
lin_reg= linear_model.LinearRegression()
lin_reg.fit(frontage_train_X, frontage_train_y)

# model results
lr_coefs = pd.Series(lin_reg.coef_,index=frontage_train_X.columns)
```

We will now use the above model results to impute the missing values in LotFrontage

```python
# use the above model predictions to populate nulls
nulls_in_lotfrontage = total_df.LotFrontage.isnull()
features = total_df_dummy[nulls_in_lotfrontage].drop('LotFrontage',axis=1)
target = lin_reg.predict(features)

# fill missing values
total_df.loc[nulls_in_lotfrontage,'LotFrontage'] = target
```

### Some basic Feature Manipulation

```python
# Since OverallQual is one of the highest correlated variables with the response, we can create a quadratic variable that might be a part of the regression equation
total_df["OverallQual_2"] = total_df["OverallQual"].astype(int) ** 2
#also creating cubic
total_df["OverallQual_3"] = total_df["OverallQual"].astype(int) ** 3
# another sqrt transformation
total_df["OverallQual_sqrt"] = np.sqrt(total_df["OverallQual"].astype(int))

# OverallQual is just a categorical variable in guise of integers
# Changing OverallQual into a categorical variable
total_df['OverallQual'] = total_df['OverallQual'].astype(str)
```

Please note, whenever you have higher order terms based on a feature in your model, it is idealto include the lower order terms as well. However to address the issue of resultant multicollinearity we can mean-center or standardize these variables which we will do in a later stage in this notebook.

We had already log-transformed GrLivArea previously to normalize the distribution. We can now create some other features out of the transformed variable.

```python
# Next features: GrLivArea
total_df['GrLivArea_Band'] = pd.cut(total_df['GrLivArea'], 6,labels=["1", "2", "3","4","5","6"])
print(total_df['GrLivArea_Band'].unique())

# since these are essential categorical variables,
# let us convert them to string
total_df['GrLivArea_Band'] = total_df['GrLivArea_Band'].astype(str)

# creating the polynomial variables from here as well
total_df["GrLivArea_2"] = total_df["GrLivArea"] ** 2
#also creating cubic
total_df["GrLivArea_3"] = total_df["GrLivArea"] ** 3
# another sqrt transformation
total_df["GrLivArea_sqrt"] = np.sqrt(total_df["GrLivArea"])
```

#### Encoding/dummyfying object variables

Of all categorical variables, a few are ordinal, others are categorical.
